```python
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# Print the Linked List
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

# Function to add node in the begining
  def AddBegining(self, newdata):
    NewNode = Node(newdata)

# Update the new nodes next to val to existing node
    NewNode.nextval = self.headval
    self.headval = NewNode

# Function to remove node by key
  def RemoveNode(self, Removekey):
    headval = self.headval
    if (headval is not None):
      if (headval.dataval == Removekey):
        self.headval = headval.next
        headval = None
        return
    while (headval is not None):
      if headval.dataval == Removekey:
        break
      prev = headval
      headval = headval.nextval
    if (headval == None):
      return
    prev.nextval == headval.nextval
    headval = None

# Create linked list named Li
Li = LinkedList()

# Add node
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Februari")
print ("Sebelum dilakukan remove")
Li.listprint()
print("=======================")

# Remove node by key = "Maret"
Li.RemoveNode("Maret")
print("Setelah dilakukan remove dengan kunci 'Maret'")
Li.listprint()
```

```python
# 1. Buat double linked list dengan 3 node didalamnya, kemudian lakukan penyisipan tengah dan hapus dengan kata kunci

# Membuat Node Double Linked List
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None


class DoubleLinkedList:
    def __init__(self):
        self.headval = None

# Menampilkan isi list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

# Menambahkan node di akhir
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return
        last = self.headval
        while last.nextval:
            last = last.nextval
        last.nextval = NewNode
        NewNode.prevval = last

# Menyisipkan node di tengah
    def AddInbetween(self, mid_node, newdata):
        if mid_node is None:
            print("The mentioned node is absent")
            return
        NewNode = Node(newdata)
        NewNode.nextval = mid_node.nextval
        mid_node.nextval = NewNode
        NewNode.prevval = mid_node
        if NewNode.nextval is not None:
            NewNode.nextval.prevval = NewNode

# Menghapus node berdasarkan kata kunci
    def RemoveNode(self, Removekey):
        headval = self.headval

        # Jika node yang dihapus adalah node pertama
        if headval is not None:
            if headval.dataval == Removekey:
                self.headval = headval.nextval
                if self.headval is not None:
                    self.headval.prevval = None
                headval = None
                return

        # Mencari node yang akan dihapus
        while headval is not None:
            if headval.dataval == Removekey:
                break
            prev = headval
            headval = headval.nextval

        # Jika data tidak ditemukan
        if headval == None:
            return

        # Mengatur ulang pointer
        prev.nextval = headval.nextval
        if headval.nextval is not None:
            headval.nextval.prevval = prev
        headval = None


# Membuat Double Linked List
Li = DoubleLinkedList()

# Menambahkan 3 node
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")

print("Sebelum sisip tengah:")
Li.listprint()
print("======================")

# Penyisipan di tengah (setelah Januari)
Li.AddInbetween(Li.headval, "Tengah")

print("Setelah sisip tengah:")
Li.listprint()
print("======================")

# Menghapus node dengan kata kunci
Li.RemoveNode("Februari")

print("Setelah hapus 'Februari':")
Li.listprint()
```

```python
# 2. Buatlah circular linked list dengan 3 node didalamnya, kemudian lakukan penyisipan akhir dan hapus akhir

# Membuat Node Circular Linked List
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class LinkedList:
    def __init__(self):
        self.headval = None

# Menampilkan isi Circular Linked List
    def listprint(self):
        printval = self.headval
        if printval is None:
            return
        while True:
            print(printval.dataval)
            printval = printval.nextval
            if printval == self.headval:
                break

# Menambahkan node di akhir Circular Linked List
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)

        # Jika list masih kosong
        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
            return

        last = self.headval
        while last.nextval != self.headval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.nextval = self.headval

# Menghapus node di akhir Circular Linked List
    def RemoveEnd(self):
        if self.headval is None:
            return

        # Jika hanya ada satu node
        if self.headval.nextval == self.headval:
            self.headval = None
            return

        last = self.headval
        while last.nextval.nextval != self.headval:
            last = last.nextval

        last.nextval = self.headval


# Membuat Circular Linked List
Li = LinkedList()

# Menambahkan 3 node
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")

print("Sebelum penyisipan akhir:")
Li.listprint()
print("==========================")

# Penyisipan di akhir
Li.AddInEnd("April")

print("Setelah penyisipan akhir:")
Li.listprint()
print("==========================")

# Hapus node di akhir
Li.RemoveEnd()

print("Setelah hapus node akhir:")
Li.listprint()
```
